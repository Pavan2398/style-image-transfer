## 2. Download Pretrained Weights

Download VGG and decoder weights. Since these are large, you can either:
- Download from the original paper's Google Drive
- Train your own encoder/decoder

```python
import os
os.makedirs("checkpoints", exist_ok=True)

# Option 1: Download from Google Drive (example)
# !gdown https://drive.google.com/uc?id=YOUR_FILE_ID -O checkpoints/vgg_normalised.pth
# !gdown https://drive.google.com/uc?id=YOUR_FILE_ID -O checkpoints/decoder.pth

# Option 2: Build VGG encoder manually (for demo)
import torch
import torchvision.models as models

# Load pretrained VGG19
vgg19 = models.vgg19(pretrained=True)

# Create encoder from VGG features
class Encoder(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.features = vgg19.features
    def forward(self, x):
        return self.features(x)

encoder = Encoder()
torch.save(encoder.state_dict(), "checkpoints/vgg_normalised.pth")
print("Created encoder weights from VGG19")
```
